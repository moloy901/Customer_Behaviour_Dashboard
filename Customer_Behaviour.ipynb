```python
HD['Review Rating'] = HD.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
HD.isnull().sum()
```

```python
HD.columns=HD.columns.str.lower()
HD.columns=HD.columns.str.replace(' ','_')
HD=HD.rename(columns={'purchase_amount_(usd)':'purchase_amount'})

```

```python
HD.columns
```

```python
HD.isnull().sum()
```

```python
#Create a group of age column
labels = ['Young-Adult', 'Adult', 'Middle-Aged' ,'Senior']
HD['age_group'] = pd.qcut (HD['age'],q=4,labels = labels )
```

```python
HD[['age','age_group']].head(20)
```

```python
#Create a column purchase_frequency_days
frequency_mapping= {
    'Fortnightly': 14,
    'Weekly': 7,
     'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
HD['purchase_frequency_days'] = HD['frequency_of_purchases'].map(frequency_mapping)
```
